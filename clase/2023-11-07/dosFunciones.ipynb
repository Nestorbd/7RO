```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define la función de dos variables
def f(x, y):
    return (x**2+(y/5))

# Genera datos para x y y
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)

# Calcula los valores de z usando la función
z = f(x, y)

# Crea una figura 3D
fig = plt.figure()

# Agrega un eje 3D a la figura
ax = fig.add_subplot(111, projection='3d')

# Grafica la superficie
ax.plot_surface(x, y, z, cmap='viridis')

# Muestra el gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.axes3d import Axes3D, get_test_data
from matplotlib import cm
import numpy as np
# Figura
fig = plt.figure()

# Tomo el eje actual y defino una proyección 3D

ax=fig.add_subplot(111, projection='3d')
# Dibujo 3D
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)

# el metodo meshgrid devuelve una matriz de coordenadas
# a partir de vectores de coordendas, que usamos para
# los datos del eje Z
X, Y = np.meshgrid(X, Y)

R =np.sqrt( X**2 + Y**2)
Z = np.sin(R)

# Grafico surface en 3D
surface = ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=cm.coolwarm, linewidth=0, antialiased=False)

# Límites del eje Z
ax.set_zlim(-1.01, 1.01)
```

```python
xtrain = np.random.uniform(-5,5, (1000000,2));
ytrain = np.zeros(1000000)

for t in range(len(xtrain)):
    X=xtrain[t][0]
    Y=xtrain[t][1]
    R=np.sqrt(X**2 + Y**2)
    Z = np.sin(R)
    ytrain[t] = np.round(Z,1)
```

```python
xtest = np.random.uniform(-5,5, (200000,2));
ytest = np.zeros(200000)

for t in range(len(xtest)):
    X=xtest[t][0]
    Y=xtest[t][1]
    R=np.sqrt(X**2 + Y**2)
    Z = np.sin(R)
    ytest[t] = np.round(Z,1)
```

```python
import keras
from keras.datasets import fashion_mnist
from keras.models import Sequential
from keras.layers import Dense
import tensorflow as tf

import matplotlib.pyplot as plt
```

```python
model = keras.Sequential([
keras.layers.Dense(2, activation='sigmoid'),
keras.layers.Dense(128, activation='sigmoid'),
keras.layers.Dense(128, activation='sigmoid'),
keras.layers.Dense(64, activation='sigmoid'),
keras.layers.Dense(1, activation='linear')
])
```

```python
#model.summary()
model.compile(optimizer='adam',
loss='mse',
metrics=['accuracy'])
model.fit(xtrain, ytrain, epochs=100,validation_data=(xtest,ytest))
```

```python
predictions=model.predict(xtest)
predictions[1]
```
